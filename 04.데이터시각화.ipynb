## 2. Customization

### Axis Label, Title

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

plt.show()
```

### Ticks

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

# Ticks
plt.yticks([0, 2, 4, 6, 8, 10]
           , ['0', '2B', '4B', '6B', '8B', '10B'])

plt.show()
```

### Color

```python
values = [0,1.6,1.4,2.2,2.5,2.6,3.2,3.5,3,3.9,4.2,6]
plt.hist(values, bins = 6, color="yellow")
plt.show()
```

### Common Axis

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

plt.plot(year, pop1, 'red')
plt.plot(year, pop2, 'blue')

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

plt.show()

```

### Subplot

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

# plot1
plt.subplot(2,1,1)
plt.plot(year, pop1, 'red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop1')

# plot2
plt.subplot(2, 1, 2)
plt.plot(year, pop2, 'blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop2')

plt.show()
```

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

# plot1
plt.subplot(2, 1, 1)
plt.plot(year, pop1, 'red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop1')

# plot2
plt.subplot(2, 1, 2)
plt.plot(year, pop2, 'blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop2')

plt.tight_layout() # Or equivalently,  "plt.tight_layout()"

plt.show()
```

## 4. seaborn 패키지

seaborn 패키지는 차트를 더 다양하고 효과적으로 시각화하도록 돕는 패키지 입니다.


```python
# 패키지 로딩
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 4.1 distplot (Density & Histogram)

```python
# 데이터 불러오기
aq = pd.read_csv("https://raw.githubusercontent.com/RayleighKim/Example_datasets/master/airquality.csv")
ti =  pd.read_csv("https://raw.githubusercontent.com/RayleighKim/Example_datasets/master/titanic_simple.csv") 
```

```python
# Density plot만 그리기 위한 .kdeplot()
sns.kdeplot(aq["Ozone"])
plt.show()
```

```python
# distplot으로 densityplot과 histogram 함께 그리기
sns.distplot(aq["Ozone"])
plt.show()
```

```python
aq["Ozone"].notnull()
```

```python
aq[aq["Ozone"].notnull()]
```

```python
aq[aq["Ozone"].notnull()]['Ozone']
```

```python
# distplot으로 densityplot과 histogram 함께 그리기
sns.distplot(aq[aq["Ozone"].notnull()]["Ozone"])
plt.show()
```

```python
# distplot으로 densityplot과 histogram 함께 그리기
sns.distplot(aq[aq["Ozone"].notnull()]["Ozone"], rug=True)
plt.show()
```

### 4.2 barplot

```python
# 먼저 집계를 한 후에...
ti_2 = ti.groupby("Pclass", as_index = False)[["PassengerId"]].count()
ti_2.columns.values[1] = "PassCnt"
```

```python
sns.barplot(x="Pclass", y = "PassCnt", data=ti_2)
plt.show()
```

```python
sns.countplot(x="Pclass", data=ti, hue = "Survived")
```

### 4.3 boxplot

```python
sns.boxplot(x="Pclass", y="Age", hue="Survived", data=ti)
```

```python
#아래 Set3을 Set1, Set2 등으로 변경해서 실행해보세요.
sns.boxplot(y="Pclass", x="Age", hue="Survived", data=ti, orient = 'h', palette = "Set3")
```

### 4.4 Scatterplot

### 4.5 특별한 차트

#### jointplot

```python
sns.jointplot(x="Fare", y="Age", data=ti)
```

#### pairplot

pairplot은 NaN이 포함된 변수를 사용할 수 없기에, 사전에 제거한 후에 사용합니다.


```python
# NaN을 제외한 데이터프레임으로 살펴보기
ti2 = ti[ti['Age'].notnull()]
aq2 = aq[aq['Ozone'].notnull()]
```

```python
sns.pairplot(aq2, vars = ['Ozone','Wind','Temp'])
```

```python
sns.pairplot(ti2, vars = ['Age','Fare'], hue = "Survived")
```
